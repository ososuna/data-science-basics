# Derivatives in Activation Functions

An activation function is a mathematical function used in artificial neural networks to determine whether a neuron should be activated or not ‚Äî essentially, it decides if the information the neuron is processing is important enough to pass on to the next layer.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Discrete derivative

```python
x = np.linspace(-10,10, num=1000)

def df(f):
  h=0.000001
  return (f(x+h)-f(x))/h
```

## Heaviside function

$$H(x) = 
     \begin{cases}
        0, &\quad \text{for, } x < 0 \\
        1,  &\quad\text{for. } x \ge 0 \\
     \end{cases}
$$

### Derivative
This function tends to infinity as x tends to zero (in this case), and equals zero for all other values.


$$f¬¥(x)=0$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N=1000

y = H(x)

plt.plot(x, H(x), 'b')
plt.plot(x,df(H), 'r')
plt.grid()
```

## Sigmoid function

$$f(x)=\frac{1}{1-e^{-x}}$$

### Derivative
Optimizing this function in neural networks can lead to a problem known as vanishing gradient, due to the complexity of the function.
$$f¬¥(x)=f(x)(1-f(x))$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
   
plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Hyperbolic tangent function

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

### Derivative
Something similar happens with this function as with the sigmoid. The complexity of its derivative can cause problems during optimization.
$$f¬¥(x)=1-f(x)^2$$

```python
def f(x):
  return np.tanh(x)
    
N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## ReLu function

$$R(x)=max(0,x)$$
 It can also be written as
$$R(x) = 
     \begin{cases}
        0, &\quad \text{for, } x < 0 \\
        x,  &\quad\text{for. } x \ge 0 \\
     \end{cases}
$$

### Derivative

$$H(x) = 
     \begin{cases}
        0, &\quad \text{for, } x < 0 \\
        1,  &\quad\text{for. } x \ge 0 \\
     \end{cases}
$$

```python
def f(x):
  return np.maximum(x,0)

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

ReLu has no derivative at zero.
