# Dispersion diagrams

## Dependencies

```python
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

## Scatterplot
http://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

## Join plot

http://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

## Box plot

http://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sns.boxplot(x='species', y='sepal_length', data=iris, estimator=mean)
```

## Bar plot

http://seaborn.pydata.org/generated/seaborn.barplot.html

```python
sns.barplot(x='species', y='sepal_length', data=iris)
```
