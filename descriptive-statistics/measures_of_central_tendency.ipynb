# Measures of Central Tendency

## üìê Mean
The **mean** (or arithmetic average) of a set of $n$ numbers $x_1, x_2, \dots, x_n$ is defined as:
$$
\text{mean} = \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

---

## üîó Median
The **median** is the middle value of the ordered data set.

- If $n$ (the number of observations) is **odd**, the median is:
$$
\text{median} = x_{\frac{n+1}{2}}
$$

- If $n$ is **even**, the median is the average of the two middle values:
$$
\text{median} = \frac{x_{\frac{n}{2}} + x_{\frac{n}{2} + 1}}{2}
$$

---

## üé≤ Mode
The **mode** is the value (or values) that appear most frequently in the data set.  
If all values occur with the same frequency, the data set is said to have **no mode**.
