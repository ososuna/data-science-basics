## Heaviside function
It "turns on" when the input is zero or more. That's why it's also called a **binary step function**.
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$
### ‚úÖ Example: Basic Binary Classifier

Imagine a simple model that checks if someone is old enough to vote (age ‚â• 18):

- Input: `age`
- Output:  
  - `0` ‚Üí "Cannot vote"  
  - `1` ‚Üí "Can vote"

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N=1000

y = H(x)

plt.plot(x,y)
plt.grid()
```

## Sigmoid function
The **sigmoid function** squashes any real-valued input into a value between **0 and 1**. It looks like an "S"-shaped curve (also called a logistic curve).
$$f(x)=\frac{1}{1-e^{-x}}$$

### ‚úÖ Example: Predicting If a Student Will Pass

Imagine you have a model that predicts whether a student will **pass or fail** an exam based on their study hours and attendance.

- Inputs:  
  - `Study hours`
  - `Attendance %`
- Output:  
  - **Close to 1** ‚Üí Likely to pass  
  - **Close to 0** ‚Üí Likely to fail

The sigmoid function is perfect here because it gives a smooth probability score between 0 and 1.

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Hyperbolic tangent function
The **hyperbolic tangent** (or `tanh`) is a smooth, S-shaped activation function that outputs values between **-1 and 1**.
$$f(x)=\frac{2}{1+e^{-2x}}-1$$

### ‚úÖ Example: Sentiment Analysis (Basic Idea)

Imagine you're building a model that processes words and tries to detect **sentiment** (positive or negative feeling):

- Input: Words from a sentence (encoded numerically)
- Output of hidden layer: `tanh` used to represent the "emotional charge" of words  
  - **Close to -1** ‚Üí Strongly negative sentiment  
  - **Close to 0** ‚Üí Neutral  
  - **Close to 1** ‚Üí Strongly positive

`tanh` helps because it can capture **both positive and negative signals**.

```python
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## ReLu function
**ReLU** stands for **Rectified Linear Unit**. It is one of the most commonly used activation functions in modern neural networks.
$$R(x)=max(0,x)$$
### ‚úÖ Example: Image Recognition

In a neural network that tries to classify images (e.g., recognize handwritten digits), ReLU is used in the hidden layers to introduce **non-linearity**:

- Input: Pixel values of the image
- Hidden Layer: ReLU activation  
  - Keeps positive activations  
  - Discards negative activations (sets them to 0)

This helps the model focus only on **relevant patterns**.

```python
def f(x):
  return np.maximum(x,0)

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
