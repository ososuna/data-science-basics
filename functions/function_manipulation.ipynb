# Function manipulation

## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Vertical and horizontal displacements

Being $c$ a constant greater than zero, then the graph:

- $y=f(x)+c$ shifts $c$ units up.
- $y=f(x)-c$ shifts $c$ units down.
- $y=f(x-c)$ shifts $c$ units to the right.
- $y=f(x+c)$ shifts $c$ units to the left.

```python
N = 1000

def f(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y, color='blue')
ax.plot(x, y+c, color='green', linestyle=':')
ax.plot(x, y-c, color='purple', linestyle=':')
ax.plot(x, f(x-c), color='orange', linestyle=':')
ax.plot(x, f(x+c), color='pink', linestyle=':')
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Stretches and Compressions

If $c$ is a constant greater than zero, then the graph:

- $y=c \cdot f(x)$ stretches the graph vertically by a factor of $c$.
- $y= \frac{1}{c} \cdot f(x)$ compresses the graph vertically by a factor of $c$.
- $y=f(c \cdot x)$ compresses the graph horizontally by a factor of $c$.
- $y= f(\frac{1}{c} \cdot x )$ stretches the graph horizontally by a factor of $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f(x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x, y*c, color='green', linestyle=':')
ax.plot(x, y/c, color='purple', linestyle=':')
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x, f(x*c), color='orange', linestyle=':')
ax.plot(x, f(x/c), color='pink', linestyle=':')
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Reflections

- $y=-f(x)$ reflects the graph with respect to the x-axis.
- $y=f(-x)$ reflects the graph with respect to the y-axis.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
