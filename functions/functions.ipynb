## Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Algebraic functions

### Linear function

Has the form of $$f(x)=mx + b$$ where $m$ and $b$  $\in R$. 

$m$ can be calculated by: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

and $b$ is the intersection point with the $y$. Its domain is $Dom_{f} = (-\infty, \infty)$. Its image is $Im_{f} = (-\infty, \infty)$




```python
N = 100
m = -1
b = 3
def f(x):
    return m*x+b
x = np.linspace(-10,10,num=N)
y = f(x)
fig, ax, = plt.subplots()
ax.plot(x,y)
ax.grid()
```

### Polinomic functions

Have the form of $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a function that has this form is called a polynomial of degree $n$. To the elements $a$ we will call them coefficients where $a \in R$. 

**For example:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

is a seventh-degree polynomial

#### Power functions

There are some functions that are a particular case of polynomial functions, which are power functions, which have the form:

$$f(x)= x^{a}, a \in R$$ 


**For example:**

$$f(x)= x^{2}$$

The domain of $f(x)=x^{2}$ is $Dom_{f} = (-\infty, \infty)$. Its image is $Im_{f} = [0, \infty)$

#### Odd-degree function

```python
def f(x):
    return x**7

y = f(x)

plt.plot(x,y)
plt.grid()
```

#### Even-degree function

```python
def f(x):
    return x**4

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Transcendental functions
Cannot be expressed as a finite combination of the basic algebraic operations (addition, subtraction, multiplication, division, and taking roots) involving polynomials.

#### Trigonometric functions
Some examples are $cos(x)$, $sen(x)$ and $tan(x)$ functions

```python
def f(x):
  return np.cos(x)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
def f(x):
  return np.sin(x)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```python
# Define the x range avoiding vertical asymptotes (odd multiples of pi/2)
xtan = np.linspace(-2 * np.pi, 2 * np.pi, N)
y = np.tan(xtan)

# Mask the values where tangent has vertical asymptotes
y[np.abs(y) > 10] = np.nan  # Avoid plotting large values for better visuals

plt.plot(x, y)
plt.grid()
```

#### Exponential function

Has the form of $$f(x)=a^x$$ where the base $a$ is a positive constant. A great example of an exponential function is using the base as Euler's number:

$$f(x)=e^x$$

```python
def f(x):
  return np.exp(x)

y=f(x)

plt.plot(x,y)
```

#### Logaritmic function

The logarithm is defined by the **relation**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

where:
* $b$ is the base.
* $n$ is the exponent to which the base is raised.
* $x$ is the result of raising the base $b$ to the exponent $n$

**For example:**

Having b=2 y n=8, then:

$$2^8=256$$

So $x=256$. Calculating the base 2 logarithm of $x$ is:

$$log_{2}(256) = 8$$

```python
def f(x):
    return np.log2(x)
xlog = np.linspace(0.001,256, num=1000)
plt.plot(xlog, f(xlog))
plt.grid()
```
