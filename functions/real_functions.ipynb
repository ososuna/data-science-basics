# Real Functions
They are called real functions because both their domain and the codomain (remember that the codomain can also be called range or image) are contained in the set of real numbers.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Even function
A function is even if it satisfies the following relationship throughout its domain:
$$f(-x) = f(x)$$
This relationship tells us that a function is even if it is symmetrical to the vertical axis (Y axis). For example, a parabola is an even function.

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return x**2
y = f(x)
plt.plot(x, y)
plt.grid()
```

## Odd function
A function is odd if it satisfies the following relationship throughout its domain:
$$f(-x) = -f(x)$$
This relationship tells us that a function is odd if it is symmetrical to the horizontal axis (X axis). For example, a cubic function is an odd function.

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return x**3
y = f(x)
plt.plot(x, y)
plt.grid()
```

## Bounded function
A function is bounded if its codomain (also known as range or image) lies between two values, i.e., it is bounded. This definition states that there is a number m such that, for every value of the function's domain, the following holds:
$$ -m \leq f(x) \leq m $$
For example, the sine or cosine function is bounded in the interval [-1, 1] within its codomain.$$

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return np.cos(x)
y = f(x)
plt.plot(x, y)
plt.grid()
```

## Monotonic Functions
These functions are useful to recognize or analyze because they allow us to determine whether a function increases or decreases in any of its intervals. Monotonic means that it has no variations. Therefore, monotonic functions are those that satisfy one of these properties within an interval I of the real numbers:

#### 1. The function is monotonic and strictly increasing:
$$\forall x_1, x_2 \in D,\ x_1 < x_2 \Rightarrow f(x_1) < f(x_2)$$
What this definition tells us is that x1 always has to be less than x2 in our interval I, and that when evaluating x2 in the function the result of this will always be greater than if we evaluate the function at x1. For example, the linear function.

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return 2*x+1
y = f(x)
plt.plot(x, y)
plt.grid()
```

#### 2. The function is monotonic and strictly decreasing:
$$\forall x_1, x_2 \in D,\ x_1 < x_2 \Rightarrow f(x_1) > f(x_2)$$

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return -2*x+1
y = f(x)
plt.plot(x, y)
plt.grid()
```

#### 3. The function is monotonic and increasing:
$$\forall x_1, x_2 \in D,\ x_1 < x_2 \Rightarrow f(x_1) \leq f(x_2)$$

```python
N = 1000
x = np.linspace(0, 10, num=N)
def f(x):
    return x**2
y = f(x)
plt.plot(x, y)
plt.grid()
```

#### 4. The function is monotonic and decreasing:
$$\forall x_1, x_2 \in D,\ x_1 < x_2 \Rightarrow f(x_1) \geq f(x_2)$$

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return np.where(x <= 0, 4, 2 - x)
y = f(x)
plt.plot(x, y)
plt.grid()
```

## Periodic functions
Periodic functions are those that repeat every certain period; this period is designated by the letter T. The relationship that the function must fulfill to be periodic is the following.
$$f(x)=f(x+T), T \neq 0$$
For example, the sine and cosine functions are periodic functions with a period T = 2. That is, if we calculate f(x) and f(x + 2) in the sine function, the value given by both expressions is the same.

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return np.sin(x+2*np.pi)
y = f(x)
plt.plot(x, y)
plt.plot(x, np.sin(x), color='red', linestyle=':')
plt.grid()
```

## Concave and convex functions
Let \( f(x) \) be a twice-differentiable function.

The **concavity** or **convexity** of \( f(x) \) at a point \( x \) is determined by the second derivative \( f''(x) \):

\begin{cases}
f''(x) > 0 & \Rightarrow \text{the function is convex (concave up) at } x \\
f''(x) < 0 & \Rightarrow \text{the function is concave (concave down) at } x \\
f''(x) = 0 & \Rightarrow \text{possible inflection point or change in concavity}
\end{cases}

### Convex
A function within an interval is said to be convex if the function "opens upward." That is, if it looks like this:

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return x**2
y = f(x)
plt.plot(x, y)
plt.grid()
```

### Concave
A function within an interval is said to be concave if the function "opens downward." That is, if it looks like this:

```python
N = 1000
x = np.linspace(-10, 10, num=N)
def f(x):
    return -(x**2)
y = f(x)
plt.plot(x, y)
plt.grid()
```
